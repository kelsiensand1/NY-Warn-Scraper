```python
import pandas as pd
import numpy as np
#import pandas and np
```

```python
df = pd.read_csv('../data/warn.csv')
#imported the csv from the scrape
```

```python
df_NY= df[df['county'] == 'New York County']
#saved a data frame just for New York County
```

```python
df_NY['business_type'].unique()
#looked at business types in the data that are unique
```

```python
df_rest= df_NY[df_NY['business_type'] == 'Restaurant'] 
df_rest
#saved a dataframe with restaurants in New York county
```

```python
df_NY.loc[df_NY['business_type'].str.contains('Restaurant|restaurant|Restaurants|restaurants'), 'business_type'] = 'Restaurant'
#contained all the NY county restaurants with spelling variations
```

```python
df_newrest = df_NY['business_type'].value_counts().reset_index().rename(columns={'index': 'business_type', 'business_type': 'count'})
df_newrest
#changed all column values which contained the word restaurants and counted it using value_count, renamed columns
```

```python
df_allrests=df_NY[df_NY['business_type'].str.contains('Restaurant|restaurant')]
#saved all restaurants variations into new variable
```

```python
df_allrests
```

```python
df_allrests['affected'].sum()
#all the people affected by restaurant layoffs
```

```python
df_allrests.tail()
```

```python
df_NY.groupby(['business_type'])['affected'].agg('sum').sort_values(ascending=False).reset_index()
#Looked at the people affected by layoffs by each business type
```
